```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
baseurl = "https://www.carsome.id"
merek = "toyota"
```

```python
productlink = []

for i in range(1,3):
    k = requests.get("https://www.carsome.id/beli-mobil-bekas/{}?pageNo={}".format(merek, i)).text
    soup = BeautifulSoup(k,'html.parser')
    productlist = soup.find_all("div",{"class":"list-card__item"})

    for product in productlist:
        link = product.find("a", {"class": "mod-b-card__title"})
        if link:
            link = link.get('href')
            productlink.append(baseurl+link)
```

```python
productlink
```

```python
data = []
```

```python

for link in productlink:
    r = requests.get(link).text
    soup = BeautifulSoup(r, 'html.parser')
    
    dat1 = soup.find("div", {"class":"car-info-top"}).text
    dat2 = soup.find("div", {"class":"car-info-bottom"}).text.replace('\n',"")
    
    if dat1:
        name_year = dat1.split(" ")
    
    if dat2:
        trans = dat2.split(" ")
    
    try:
        name = name_year[2]
    except:
        name = None
    
    try:
        brand = name_year[1]
    except:
        brand = None
    
    try:
        year = name_year[0]
    except:
        year = None
        
    try:
        tipe = name_year[3]+" "+name_year[4]
    except:
        tipe = None
        
    try:
        distance = trans[6]
    except:
        distance = None
        
    try:
        transmisi = trans[19]
    except:
        transmisi = None
    
    try:
        price = soup.find("div", {"class":"price"}).text.replace('\n',"")
    except:
        price = None
        
    product = {
        "name": name,
        "brand": brand,
        "tipe": tipe,
        "year": year,
        "distance": distance,
        "transmisi": transmisi,
        "price": price,
    }
    
    data.append(product)
```

```python
data
```

```python
df = pd.DataFrame(data)
```

```python
df.to_csv("data_mobil.csv")
```
